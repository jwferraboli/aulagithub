## 1 – Um objeto composto multidimensional (matriz) pode receber  objeto composto unidimensional (vetor)?

```python
#Sim, uma matriz pode receber um vetor
##Desde que: inserido como linha ou coluna indexando da matriz (indexando) e que tenha mesmo formato de tamanho
```

```python
##Instanciando uma matriz do tipo lista
matriz = [
    [1, 2, 3], 
    [4, 5, 6]
    ]

vetor = [7, 8, 9]
matriz[1] = vetor

print(matriz)
```

## 10 – Importe o data.csv (dados de seguro na Flórida em 2012) para uma DataFrame. Armazene em um dicionário o número de entradas, uma lista com as classes das colunas e outra lista com o número de elementos de cada classe, três listas, contendo a média, o desvio padrão e o terceiro quartil da primeira, quarta e nona colunas . Recupere o dicionário.

```python
import pandas as pd
```

```python
dfseguro = pd.read_csv("data.csv")
dfseguro.head()
```

```python
entradas = dfseguro.shape
entradas
```

```python
classes = [dfseguro.dtypes.unique()]
classes
```

```python
numero_classes = [dfseguro.dtypes.value_counts()]
numero_classes
```

```python
colunas = dfseguro.iloc[:, [0, 3, 8]]
colunas
```

```python
listamedia = [colunas.mean]
listamedia
```

```python
listadesvio = [colunas.std]
listadesvio
```

```python
quartil3 = [colunas.quantile(0.75)]
quartil3
```

```python
dicionario_seguro = {
    "entrada": entradas,
    "classes": classes,
    "num_classes": numero_classes,
    "media": listamedia,
    "desvio_pad": listadesvio,
    "quartil": quartil3
}
dicionario_seguro
```
