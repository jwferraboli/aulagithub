```python
maiores_20 = obj_array > 20
print(maiores_20)
```

## 09 – Escreva o programa em Python:

```python
import pandas as pd
df_novo10 = pd.read_csv("arquivoaula10.csv", index_col="Unnamed: 0")
df_novo10
```

```python
## Usando o número de caracteres do primeiro gene, quero saber a informação gênica
gapdh = df_novo10.iloc[0]
print(gapdh)
```

```python
##Caso o nome tenha entre 2 e 4 caracteres inclusive, posso saber as informações do gene?
minimo = 2
maximo = 4
if minimo < gapdh["n_caracteres"] <= maximo:
    print("Informação do gene:")
    print(gapdh)
```

```python
#Caso tenha menos de 2 caracteres, posso saber todos os valores nas instâncias onde os números de caracteres dos nomes são superiores a 4?

if minimo < gapdh["n_caracteres"] <= maximo:
    print("Informação do gene:")
    print(gapdh)

elif gapdh["n_caracteres"] < minimo:
    print("Valores nas instâncias:")
    display(df_novo10[df_novo10["n_caracteres"] > 4])
```

```python
# Caso tenha mais de 4 caracteres, apenas me informe quem é superior.

if minimo < gapdh["n_caracteres"] <= maximo:
    print("Informação do gene:")
    print(gapdh)
    
elif gapdh["n_caracteres"] < minimo:
    print("Valores nas instâncias:")
    display(df_novo10[df_novo10["n_caracteres"] > 4])

else:
    print("Quem é superior:")
    display(df_novo10[df_novo10["n_caracteres"] > 4].index)
```
