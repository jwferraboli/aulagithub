## 05 – Seria possível, com o racional proposto, saber quantos e quais valores são superiores a 20? Porque?

```python
maiores_20 = obj_array > 20
print(maiores_20)
```

```python
##Quantos com count
maiores_20 = list(obj_array > 20).count(True)
print(maiores_20)
```

```python
##Quantos com .sum()
quantos_20 = (obj_array > 20).sum()
print(quantos_20)
```
