```python
#Compara os valores, se o primeiro valor é máximo em relação a outros valores
id_max = df["DV1 BR90"] == df["DV1 BR90"].max()
id_max.head()
```

```python
def highlight_max(s):
    is_max = s == s.max()
    return ['backgroud-color: yellow' if v else ' ' for v in is_max]
```

```python
##Deveria ficar colorido
(
   df[df["Sample"] == "Sample 1"]
   .style
   .apply(highlight_max, axis=0)
)
```
