```python
## Incluindo função
def gerar_obj12():
    return np.random.normal(10, 3, 5)
```

```python
def valores_maiores_que_10(array, patinho_amarelo: int = 10):
    quais = array[array > patinho_amarelo]
    quantos = len(quais)
    return quais, quantos
```

```python
obj = gerar_obj12()
quais, quantos = valores_maiores_que_10(obj)

print("Array gerado:", obj)
print("Quais valores > 10:", quais)
print("Quantidade de valores > 10:", quantos)
```

R2 =define a media de valores, na qual a soma recebe o comando sum contido dentro de valores; quantidade recebe o comando len para a quantidade de valores dentro do conjunto valores; média recebe soma/quantidade, retornar a média

## 07 – Escreva o programa em Python:

```python
n_valores = [10, 24, 33, 45, 2]
resultado = media(n_valores)

def media(valores):
    soma = sum(valores)
    quantidade = len(valores)
    media = soma / quantidade
    return media

print("Média:", resultado)
```

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
obj1 = 10
obj2 = 5

def dividir(obj1, obj2):
    if type(obj1) == int and type(obj2) == int:
            return obj1/obj2
    else:
            return "Valores não são inteiros"

print("obj1:", type(obj1), "valor:", obj1)
print("obj2:", type(obj2), "valor:", obj2)
print("resultado:", dividir(obj1, obj2))
```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
obj1 = (1, 2, 10)
obj2 = (25, 42, 2)

def dividir(obj1, obj2):
    if type(obj1) == int and type(obj2) == int:
            return obj1/obj2
    else:
            return "Valores não são inteiros"

print("obj1:", type(obj1), "valor:", obj1)
print("obj2:", type(obj2), "valor:", obj2)
print("resultado:", dividir(obj1, obj2))
```
