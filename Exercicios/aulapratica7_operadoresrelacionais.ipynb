```python
## conjunto de dados aleatórios iguais
np.random.seed(42)
array_1 = array = np.random.normal(0, 1, 1_000).reshape(-1, 10)
array_2 = array = np.random.normal(0, 1, 1_000).reshape(-1, 10)
print(array_1[:1])
print(array_2[:1])
```

```python
array_1 == array_2
```

```python
pd.DataFrame(array_1) <= pd.DataFrame(array_2)
```

```python
filtro_array = array_1 <= array_2
filtro_array[:2]
```

```python
df_1 = pd.DataFrame(array_1)
df_2 = pd.DataFrame(array_2)
df_1.head()
```
