R7 = Importa a biblioteca de visualização gráfica e define um gráfico de barras (bar), no qual o eixo x representa o número de caracteres e o eixo y representa a frequência. Por fim, o gráfico é exibido.

## 06 – Escreva o programa em Python:

```python
import pandas as pd
df_novo13 = pd.read_csv("arquivoaula10.csv", index_col="Unnamed: 0")
display(df_novo13)
```

```python
freq_abs = df_novo13["n_caracteres"].value_counts().sort_index()
freq_abs
```

```python
freq_rel = freq_abs / freq_abs.sum()
freq_rel
```

```python
freq_acum = freq_abs.cumsum()
freq_acum
```

```python
freq_acum_relativa = freq_acum.cumsum()
freq_acum_relativa
```

```python
df_novo13["freq_abs"] = df_novo13["n_caracteres"].map(freq_abs)
df_novo13["freq_rel"] = df_novo13["n_caracteres"].map(freq_rel)
df_novo13["freq_acum"] = df_novo13["n_caracteres"].map(freq_acum)
df_novo13["freq_acum_relativa"] = df_novo13["n_caracteres"].map(freq_acum_relativa)
df_novo13
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(
    kind='bar',
    x=list(df_novo13["freq_abs"].index),
    y='freq_acum',
    data=df_novo13,
)
plt.xlabel("Concentrações teste")
plt.ylabel("Frequência")
plt.title("Frequências Acumulada Absoluta")
plt.show()
```

```python
sns.catplot(
    kind='bar',
    x=list(df_novo13["freq_abs"].index),
    y='freq_acum_relativa',
    data=df_novo13,
)
plt.xlabel("Concentrações teste")
plt.ylabel("Frequência")
plt.title("Frequências Acumulada Relativa")
plt.show()
```

```python
sns.catplot(
    kind='bar',
    x=list(df_novo13["freq_abs"].index),
    y='freq_rel',
    data=df_novo13,
)
plt.xlabel("Concentrações teste")
plt.ylabel("Frequência")
plt.title("Frequências Relativas Simples")
plt.show()
```

## 09 – Repita o exercício 8, inserindo na DataFrame as replicas "Amostra 3“, "Amostra 4“, "Amostra 5“ e "Amostra 6".

```python
def magica(numero: str) -> pd.DataFrame: 
    
    objeto_verde = np.random.randint(1, 10, 5)
    objeto_verde = pd.DataFrame({
        "Frequência Simples Absoluta": objeto_verde,
        "Frequência Simples Relativa": objeto_verde/objeto_verde.sum()
        
    })
    objeto_verde.index= df_novo14.index
    objeto_verde["Frequência Acumulada Absoluta"]= objeto_verde.loc[:,"Frequência Simples Absoluta"].cumsum()
    objeto_verde["Frequência Acumulada Relativa"]= objeto_verde.loc[:,"Frequência Simples Relativa"].cumsum()
    
    objeto_verde["Amostras"] = f"Bolinha de queijo {numero}"
    return objeto_verde 
```

```python
objeto_verde2 = magica(2)
objeto_verde3 = magica(3)
objeto_verde4 = magica(4)
objeto_verde5 = magica(5)
objeto_verde6 = magica(6)
display(objeto_verde3)
```

## 10 – Obtenha dos seus dados o gráfico abaixo, adicionando 3 replicas.

```python
replicar1 = pd.concat([
  df_novo14, 
  objeto_verde, 
  objeto_verde2, 
  objeto_verde3, 
  objeto_verde4, 
  objeto_verde5, 
  objeto_verde6],
  axis=0
)
replicar1["Replicas"]= "replicar1"
replicar1.head()
```

```python
replicar2 = pd.concat([magica(1),
magica(2),
magica(3),
magica(4),
magica(5),
magica(6)],
axis=0)
replicar2["Replicas"]= "replicar2"
replicar2.head()
```

```python
replicar3 = pd.concat([magica(1),
magica(2),
magica(3),
magica(4),
magica(5),
magica(6)],
axis=0)
replicar3["Replicas"]= "replicar3"
replicar3.head()
```

```python
replicar_verde = pd.concat([replicar1, replicar2, replicar3])
replicar_verde
```

```python
replicar_verde = replicar_verde.reset_index()
replicar_verde 
```

```python
sns.catplot(
    kind='bar',
    x='Amostras',
    y='Frequência Simples Absoluta',
    hue='index',
    height=10,
    aspect=3,
    data=replicar_verde,
)
plt.xlabel("Concentrações teste")
plt.ylabel("Frequência Absoluta Simples")
plt.title("Frequências Absoluta Simples por Teste e Réplica")
plt.show()
```
