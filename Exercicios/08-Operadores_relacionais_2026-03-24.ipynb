## 03 – Usando o método DataFrame.to_csv(‘nome_do_arquivo.csv’) exporte o objeto obtido no exercício 10 da aula anterior para um arquivo CSV. Arredonde os valores para até três casas decimais, usando o método round().
Recupere a DataFrame.

```python
import pandas as pd
```

```python
df_novo7 = pd.read_csv("arquivoaula7.csv", index_col="Unnamed: 0")
df_novo7 = df_novo7.round(3)

df_novo7
```

## 04 – Instancie um numpy array que apresente o mesmo formato (linhas e colunas) da DataFrame. Este array deve conter apenas inteiros uns.
Recupere o array.

Obs.: Procure na internet o método numpy.ones()

```python
import numpy as np
```

```python
array_novo7 = np.ones(df_novo7.shape)
print(array_novo7)
```

## 05 – Quantos dos valores contidos na DataFrame são iguais aos valores no array?

```python
df_array = df_novo7.to_numpy()
iguais = df_array == array_novo7
n_iguais = np.sum(iguais)
print(n_iguais)
```

## 06 – Quantos dos valores contidos na DataFrame são superiores aos valores no array?

```python
superior = df_array > array_novo7
n_superior = np.sum(superior)
print(n_superior)
```

## 07 – Quantos dos valores contidos na DataFrame são iguais ou inferiores aos valores no array?

## 09 – Quantos dos valores das médias das linhas são inferiores ou iguais à 5?

Obs 1.: Usando o método de classe, você pode converter qualquer objeto em outra classe. Ex. obj = list(array)

Obs 2.: Assim com o objeto da classe str, listas também possuem a função count()

```python
media_linhas = df_novo7.mean(axis=1)
lista = list(media_linhas <= 5)
n_linhas = lista.count(True)
print(n_linhas)
```

```python
print(media_linhas)
```

## 10 – Quantos dos valores das médias das colunas são superiores à 5?

```python
media_colunas = df_novo7.mean(axis=0)
lista2 = list(media_colunas > 5)
n_colunas = lista2.count(True)
print(n_colunas)
```

```python
print(media_colunas)
```
